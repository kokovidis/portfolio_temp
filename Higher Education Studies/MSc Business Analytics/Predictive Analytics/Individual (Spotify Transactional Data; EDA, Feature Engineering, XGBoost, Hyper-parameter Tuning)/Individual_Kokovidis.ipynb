## 0. Business Case Understanding

### INTRODUCTION

Over the last few years, the music industry has been dominated by digital streaming services, which produce vast amounts of data on listeners and their preferences. 

This has required major players in the industry to adopt a data driven approach to content delivery in order to stay competitive. 

Warner Music Group is looking to leverage its rich database to better understand the factors that have the most significant impact on the success of a new artist. This will allow them to optimize the allocation of resources when signing and promoting new artists.

Warner’s (large) database contains several sources of data, including the streaming platforms Spotify, Amazon Live and Apple Music. 

For this case study, we will be looking using the Spotify dataset to predict the success of artists. In particular, we want to understand the role of Spotify playlists on the performance of artist.

### Streaming Music

When artists release music digitally, details of how their music is streamed can be closely monitored. 

Some of these details include:

- How listeners found their music (a recommendation, a playlist)
- Where and when (a routine visit to the gym, a party, while working). 
- On what device (mobile / PC)
- And so on…

Spotify alone *process nearly 1 billion streams every day* (Dredge, 2015) and this streaming data is documented in detail every time a user accesses the platform. 

Analyzing this data potentially enables us to gain a much deeper insight into customers’ listening behavior and individual tastes. 

Spotify uses it to drive their recommender systems – these tailor and individualize content as well as helping the artists reach wider and more relevant audiences. 

Warner Music would like to use it to better understand the factors that influence the *future success of its artists*, *identify potentially successful acts* early on in their careers and use this analysis to make resource decisions about how they market and support their artists.

### What are Spotify Playlists and why are relevant today?

### BUSINESS PROBLEM → DATA PROBLEM

Now that we have a better understanding of the business problem, we can begin to think about how we could model this problem using data. 

The first thing we can do is defining a criterion for measuring artist success. 

Based on our business problem, one way in which we can do this is to create a binary variable representing the success / failure of an artist and determined by whether a song ends up on a key playlist (1), or not (0). We can then generate features for that artist to determine the impact they have on the success of an artist.

Our problem thus becomes a classification task, which can be modeled as follows:

### *Artist Feature 1 + Artist Feature 2 …. + Artist Feature N = Probability of Success*

```python
# Define Dependent Variable

def get_successful_artists(DF, key_filter):
    assert isinstance(DF, pd.core.frame.DataFrame), 'Argument must be a pandas DataFrame'
    if not isinstance(DF['DateTime'].dtype, np.datetime64): 
        DF['DateTime'] = pd.to_datetime(DF.DateTime)
    return DF.loc[((key_filter) & (DF.DateTime >= pd.Timestamp(2017,1,1)) ), 'artist_name' ].unique()

    #Create label in DF which shows if an artist featured in top 4 playlists in 2017
def create_successful_artists_labels(DF, key_filter): #it is not a global function - it assumes that success_2017 it's the response
    #returns the name of artists that they featured in top 4 at any time in 2017
    DF['success_2017'] = 0
    #featured_artists_2017 = DF.loc[((key_filter) & (DF.DateTime >= pd.Timestamp(2017,1,1)) ), 'artist_name' ].unique()
    DF.loc[data.artist_name.isin(get_successful_artists(DF, key_filter)), 'success_2017']=1
    return DF  

def get_successful_before_2017(DF, key_filter):
    assert isinstance(DF, pd.core.frame.DataFrame), 'Argument must be a pandas DataFrame'
    if not isinstance(DF['DateTime'].dtype, np.datetime64): 
        DF['DateTime'] = pd.to_datetime(DF.DateTime) 
    return DF.loc[((key_filter) & ( DF.DateTime <= pd.Timestamp(2016,12,31)) ), 'artist_name' ].unique()

    #Create label in DF which shows if an artist featured in top 4 playlists before 2017 (2014-2016)
def create_successful_before_2017_labels(DF, key_filter): #it is not a global function - it assumes that success_2017 it's the response
    #returns the name of artists that they featured in top 4 at any time before 2017
    DF['success_before_2017'] = 0
    DF.loc[DF.artist_name.isin(get_successful_before_2017(DF, key_filter)), 'success_before_2017'] = 1
    return DF 
```

<b> 3.5.2 Other Artist Features</b> 

<p>Below we calculate the percentage of streams from premium account per artist. Next, with artists_streamlength_median_and_mean_ratio(trans_DF, feat_DF) we calculate the median stream length (time) per artist but also a ratio. This ratio is based on the following: 
    
  <li>We calculate the median and the max stream_length per track of artist</li>
  <li>We calculate a ration of median/max(stream_length) per track</li>
  <li>Finally, wget the mean of median/max(stream_length) for each artist</li>
</ol>

```python
def artists_premium_streams_percentage(trans_DF, feat_DF):  #pct of streams from premium accounts
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'

    temp = trans_DF.groupby(['artist_name', 'access']).agg({'customer_id': 'count'})
    temp = temp.groupby(level=0).apply(lambda x:  x / float(x.sum()))
    temp = temp.reset_index()
    temp = temp.loc[temp.access=='premium']
    del temp['access']
    temp = temp.set_index('artist_name')
    feat_DF['art_premium_streams_pct'] = temp
    feat_temp = feat_DF
    return feat_temp

#artists_premium_streams_percentage(trans_pred, feat_pred)
```

```python
def artists_streamlength_median_and_mean_ratio(trans_DF, feat_DF):  #pct of streams from premium accounts
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'

    temp = trans_DF.groupby(['artist_name', 'track_name'])['stream_length'].agg({'max' ,np.median})
    temp['track_streamlength_ratio']=temp['median']/temp['max']
    temp_art1=temp.groupby('artist_name')['median'].agg(np.median).to_frame('art_median_streamlength').reset_index()
    temp_art2 = temp.reset_index().groupby('artist_name')['track_streamlength_ratio'].mean().to_frame('art_streamlength_ratio_mean').reset_index()
    
    feat_temp = feat_DF.reset_index().copy()
    feat_temp = feat_temp.merge(temp_art1, on='artist_name', how='left').merge(temp_art2, on='artist_name', how='left').set_index('artist_name')
    return feat_temp

#artists_streamlength_median_and_mean_ratio(trans_pred, feat_pred)
```

## Playlist Features

Understanding an artist’s growth as a function of his/her movement across different playlists is potentially key to understanding how to identify and breakout new artists on Spotify. 

In turn, this could help us identify the most influential playlists and the reasons for their influence.

One way to model the effect of playlists on an artist’s performance has been to include them as categorical features in our model, to note if there are any particular playlists or combinations of playlists that are responsible for propelling an artist to future success:

### *Artist Feature 1 + Artist Feature 2 …. + Artist Feature N = Probability of Success*
**
Success (1) = Artist Features on Key Playlist
Failure (0) = Artist Not Featured on Key Playlist
**

Where,

**
⇒Artist Feature N = Prior Playlist 1 + Prior Playlist 2 +…Prior Playlist N
**


Given that we have over 19,000 playlists in our dataset or 600 artists, using the playlists each artist has featured on, as categorical variables would lead to too many features and a very large, sparse matrix. 

Instead, we need to think of ways to summarize the impact of these playlists. One way to do this would be to consider the top 20 playlists each artist has featured on. 

Even better would be to come up with one metric that captures the net effect of all top 20 prior playlists, for each artist, rather including using all 20 playlists for each artists as binary variables. The intuition here is that if this metric as a whole has an influence on the performance of an artist, it would suggest that rather than the individual playlists themselves, it is a combination of their generalized features that affects the future performance of an artist. 

Accordingly, different combinations of playlists could equate to having the same impact on an artist, thereby allowing us to identify undervalued playlists.

Some of the features such a metric could use is the number of unique users or ‘reach’, number of stream counts, and the passion score of each playlist

* Prior Playlist Stream Counts
* Prior Playlist Unique Users (Reach)
* Prior Playlist Passion Score

There are several other such features that you could generate to better capture the general characteristics of playlists, such as the average lift in stream counts and users they generate for artists that have featured on them.

<b> 3.6.1 Playlist features; Ranking for most streamed playlists per artist</b> 

<p>Below we demonstrate how we can create a ranking in transcactions' DataFrame which shows in descending order the numbers of streams of artist per playlist. To achieve this we use the .rank(method='first', ascending=False) method. In this way the playlist with the most streams of any given artist will have a rank=1, the second rank=2 and so on. In this way we can keep the top 20 playlist of any given artist by filtering for rank>20.</p>
    
</div>

```python
def play_inverse_ranking_by_total_streams(trans_DF):
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert 'rank_first' not in trans_DF.columns ,'Ranking has been created already'
    #trans_DF_x = trans_DF.copy(deep=True) #deep=True
    
    ranking = trans_DF.groupby(['artist_name','playlist_name'])['customer_id'].count().to_frame('play_streams_count').reset_index()
    ranking['rank_first'] = ranking.groupby(['artist_name'])['play_streams_count'].rank(method='first', ascending=False)
    ranking = ranking.sort_values(by='play_streams_count', ascending=False)
    temp = trans_DF.merge(ranking , on=['artist_name', 'playlist_name'], how='left')
    return temp

#trans_pred = play_inverse_ranking_by_total_streams(trans_pred)

#Sanity check
#trans_pred.loc[trans_pred['artist_name']=='Maggie Lindemann', ['artist_name','play_streams_count', 'rank' ]].sort_values(by='rank', ascending=True).drop_duplicates()
```

<b> 3.6.2 Playlist features using our rank</b> 

<p>Below we define functions for creating four different features:
   <li>The first one keeps the total number of streams for an artist in his or her top 20 playlists. </li>
  <li>The second the average number of unique users who are in the top 20 playlists of each artist. </li>
  <li>The third a passion score which derives from the previous two. </li>
      <li>The fourth an alternative version of first one which keeps the average number of all streams (including streams to other artists) for the 20 top playlists of each artist. </li>
</ol>

```python
#avg number of streams of artist in its 20 top playlists
def play_total_streams_top_20(trans_DF, feat_DF):
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'
    feat_DF['play_streams_count_top_20'] = trans_DF.loc[trans_DF['rank_first']<=20].groupby('artist_name')['play_streams_count'].count()
    return feat_DF

def play_avg_number_of_users_top_20(trans_DF, feat_DF):
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'
    
    stream_20_count = trans_DF.groupby(['playlist_name'])['customer_id'].nunique().to_frame('play_user_count').reset_index()
    temp = trans_DF.loc[trans_DF['rank_first']<=20].merge(stream_20_count, on='playlist_name', how='left')
    feat_DF['play_users_avg_top_20'] = temp.groupby('artist_name').agg('mean')['play_user_count']
    return feat_DF

def play_avg_passion_score(feat_DF):
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'DataFrame must be artist_name-indexed'
    feat_DF['play_passion_score'] =  feat_DF['play_streams_count_top_20'] /feat_DF['play_users_avg_top_20']
    return feat_DF

#avg number of streams out of the total number of streams in the top 20 playlists of each artist
def play_avg_number_streams_from_all_users_in_top_20(trans_DF, feat_DF):
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'
    
    stream_20_count = trans_DF.groupby(['playlist_name'])['customer_id'].count().to_frame('play_all_streams_count').reset_index()
    temp = trans_DF.loc[trans_DF['rank_first']<=20].merge(stream_20_count, on='playlist_name', how='left')
    feat_DF['play_all_streams_avg_top_20'] = temp.groupby('artist_name').agg('mean')['play_all_streams_count']
    return feat_DF
```

```python
def age_breakdown(trans_DF, feat_DF):
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'
    assert 'user_age_label_senior_pct' not in trans_DF.columns, 'Function already executed'
    # Age breakdown
    age_temp = trans_DF.drop_duplicates(subset = ['customer_id'])
    
    age_temp['birth_year'] = age_temp['birth_year'].fillna(0)
    
    age_temp['age'] = (pd.to_numeric(age_temp.year) - pd.to_numeric(age_temp.birth_year)) #recalculating age based on stream year
    
    #def age_percentages(df):
    bins = [0, 18, 25, 50, 100]

    group_names = ['user_age_label_dependent_count', 'user_age_label_youngadult_count', 'user_age_label_adult_count', 'user_age_label_senior_count']
    categories = pd.cut(age_temp['age'], bins, labels=group_names)
    age_temp['age_label'] = categories
    age_temp = age_temp.loc[:,['artist_name', 'age_label']].set_index('artist_name')
    age_temp = pd.get_dummies(age_temp)
    age_temp.columns = group_names
    age_temp.reset_index(inplace=True)

    age_temp = age_temp.groupby('artist_name').agg({'user_age_label_dependent_count':'sum', 'user_age_label_youngadult_count':'sum', 'user_age_label_adult_count':'sum','user_age_label_senior_count':'sum'  })
    age_temp['Total'] = age_temp.sum(axis=1)
    age_temp['user_age_label_dependent_pct'] = age_temp['user_age_label_dependent_count']/age_temp['Total']
    age_temp['user_age_label_youngadult_pct'] = age_temp['user_age_label_youngadult_count']/age_temp['Total']
    age_temp['user_age_label_adult_pct'] = age_temp['user_age_label_adult_count']/age_temp['Total']
    age_temp['user_age_label_senior_pct'] = age_temp['user_age_label_senior_count']/age_temp['Total']
    del age_temp['Total']
    feat_temp = feat_DF.reset_index().merge(age_temp, on='artist_name', how='left').set_index('artist_name')
    return feat_temp

#TODO: update a single DF inside a function / not available https://stackoverflow.com/a/31621298

#feat_pred = age_breakdown(trans_pred, feat_pred)
```

**Principle Component Analysis**

The data also contains a partial region code of the listener. We might want to consider including the regional breakdown of streams per artist as a feature of our model, to know if streams for certain regions are particularly influential on the future performance of an artist. 

However, we have over 400 unique regions and like playlists, including them all would lead to too many features and a large sparse matrix. One way in which to extract relevant ‘generalized’ features of each region would be to incorporate census and demographic data, from publicly available datasets. 

This is however beyond the scope of this courswork. Instead, a better way to summarize the impact of regional variation in streams is to use dimensionality reduction techniques. Here we will use Principle Component Analysis (PCA) to capture the regional variation in stream count.

PCA captures the majority of variation in the original feature set and represents it as a set of new orthogonal variables. Each ‘component’ of PCA is a linear combination of every feature, i.e. playlist in the dataset. Use **`scikit-learn`**’s PCA module (Pedregosa, et al., 2011) for generating PCA components.

For a comprehensive understanding of how sklearn's PCA module works, please refer to the sklearn documentation. We will using 10 components of PCA in our model.

<b> 3.8.1 PCA Features </b> 

<p>Below we define two functions. The first one, initially creates a matrix which keeps the total number of streams per artist and region area. Based on this, the function then scales the matrix with the MinMax scaler. Finally keeps a selected number of Principal Components.In the same fashion the second function creates Principal Components which encapsulate the median stream length per artist and region area. <br> <br> Please note than we use MinMax scaler as we believe that we need to standardize our data, given that each region has different population and thus different scale of values.</p>
    
</div>

```python
# Region Code PCA
#https://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html
#https://towardsdatascience.com/scale-standardize-or-normalize-with-scikit-learn-6ccc7d176a02

def region_minmax_stream_count_pca(trans_DF, feat_DF , n_comp):
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'

    region_temp = trans_DF.groupby(['artist_name', 'region_code'])['customer_id'].count().unstack(fill_value=0).stack().to_frame('region_streams_total')
    region_temp = region_temp.unstack().fillna(0)
    region_temp_2 = region_temp.copy()
    
    from sklearn.preprocessing import MinMaxScaler
    #Min Max Scaling
    scaler = MinMaxScaler()
    scaler.fit(region_temp_2)
    region_temp_2 = pd.DataFrame(scaler.transform(region_temp_2))
    
    from sklearn import decomposition
    pca = decomposition.PCA(n_components=n_comp)
    pca_comp = pd.DataFrame(pca.fit_transform(region_temp_2))
    pca_comp.index = region_temp.index
    pca_comp = pca_comp.add_prefix('region_stream_count_pc_')
    pca_comp = pca_comp.reset_index()
    feat_temp = feat_DF
    feat_temp = feat_temp.reset_index()
    feat_temp = feat_temp.merge(pca_comp, on='artist_name', how='left')
    feat_temp = feat_temp.set_index('artist_name')
    return feat_temp
```

```python
def region_minmax_streamlength_median_pca(trans_DF, feat_DF , n_comp):
    assert isinstance(trans_DF, pd.core.frame.DataFrame), 'First argument must be a pandas DataFrame'
    assert 'customer_id' in trans_DF.columns, 'First DataFrame must have transactional Data'
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Second argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Second DataFrame must be artist_name-indexed'

    region_temp = trans_DF.groupby(['artist_name', 'region_code'])['stream_length'].median().unstack(fill_value=0).stack().to_frame('region_streams_total')
    region_temp = region_temp.unstack().fillna(0)
    region_temp_2 = region_temp.copy()
    
    #Min Max Scaling
    scaler = MinMaxScaler()
    scaler.fit(region_temp_2)
    region_temp_2 = pd.DataFrame(scaler.transform(region_temp_2))
    
    pca = decomposition.PCA(n_components=n_comp)
    pca_comp = pd.DataFrame(pca.fit_transform(region_temp_2))
    pca_comp.index = region_temp.index
    pca_comp = pca_comp.add_prefix('region_streamlength_median_pc_')
    pca_comp = pca_comp.reset_index()
    feat_temp = feat_DF
    feat_temp = feat_temp.reset_index()
    feat_temp = feat_temp.merge(pca_comp, on='artist_name', how='left')
    feat_temp = feat_temp.set_index('artist_name')
    return feat_temp
```

```python
import re
from sklearn.impute import SimpleImputer

# Handle missing values
#fill specific columns their nan with 0
def count_not_region_zero(feat_DF):
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Argument must be artist_name-indexed'

    #find columns which have "count" in them and remove columns that have "region" in them
    mylist = feat_DF.columns.values.tolist()
    r_count = re.compile(".*count")
    newlist = list(filter(r_count.match, mylist))  # Note 1
    r_region = re.compile(".*region")
    count_not_region = (set(newlist)) - set(list(filter(r_region.match, mylist)))
    
    feat_temp = feat_DF
    feat_temp.loc[:,count_not_region  ] = feat_temp.loc[:,count_not_region  ].fillna(0)
    return feat_temp

feat_pred = count_not_region_zero(feat_pred)

# Handle rest missing values with median
def impute_all_median(feat_DF):
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Argument must be artist_name-indexed'
    
    #feat_DF_index_labels = feat_DF.index
    imp = SimpleImputer(missing_values=np.nan, strategy='median')
    feat_temp = pd.DataFrame(imp.fit_transform(feat_DF), index = feat_DF.index, columns = feat_DF.columns)
    return feat_temp

feat_pred = impute_all_median(feat_pred)
```

```python
# Remove one of highly correlated varibles (test removing other as well)

def coll_comb(feat_DF):
    assert isinstance(feat_DF, pd.core.frame.DataFrame), 'Argument must be a pandas DataFrame'
    assert 'artist_name' in feat_DF.index.name, 'Argument must be artist_name-indexed'
    
    df = feat_pred.copy()
    cor = df.corr().abs()
    cor.index = feat_pred.columns.values.tolist()
    comb = cor.unstack()
    comb = comb.sort_values(kind="quicksort", ascending=False)
    comb = comb.reset_index()
    comb.columns = ['feat_1', 'feat_2', 'Corr']
    comb = comb.loc[(comb['feat_1']!=comb['feat_2']) & (comb['Corr']>0.95)]
    comb = comb.loc[::2]
    return comb

coll_comb(feat_pred)
```

In the next code block we balance our dataset with the SMOTE (Synthetic Minority Over-sampling Technique). We decided this step to succeed the train-test split as can lead to more accurate results and findings. To come to this decision we searched for best practices using SMOTE. The following crowd-based resource: https://datascience.stackexchange.com/a/15633 maybe describes the best the concept. 
    
However, SMOTE normally should be applied only in the train data as we want our test subset to contain only real observations and not synthetic ones (so we can avoid bias). In this scenario, normally we would end up with a test subset that contains only 6 succesful observations. After several trials we realised that having a small dataset can lead to non-robust results. 
    
By performing several trials with different seed numbers for train-test split but also by using Cross-Validated trainings, we realize that the classifications scores differ by large. We believe that this happens as different train-test splits assign different kind of succesful artists to train and test subsets (so the training subset it is or it is not good enough to form a model that can predict the rest artists). Ideally, to overcome this problem we would like to have more succesful artists in our original dataset.
    
In our case, we decide to use SMOTE also on the test subset, so we can generate more succesful artists for our prediction. Finally we end up with 65 positive observations (compared to 6 without SMOTE). We recognise that this is might not the best approach, although it has been proved to offer more stable classification metrics (for different train-test splits).

<b> 4.7 XGBoost with Kernel PCA (using Hyperparameter Tuning with Randomized Cross-Validated Search) </b> 
    
In this step we are going to use the preamble work for Kernel PCA from section 3.8.4 . Kernel PCA has to be fine-tuned in order to provide Principal Components which can contribute to our predictive model. Below we provide an illustration on how we are going to use Kernel PCA in our Machine Learning pipeline:
<li> First we generate the region features and we create kernel principal components for different parameters</li>
<li> Then, we merge these principal components with the rest features (artist, playlist, user features)</li>
<li> We pass these to XGBoost algorithm and we train for a different of set of parameters </li>
<li> We store the results and the parameters from both the Kernel PCA and XGBoost after each iteration</li>
<li> We retrieve the best parameters based on our scoring criterion on XGBoost (in our case is AUC).</li>
    
The pipeline is orchestrated by RandomizedSearch CV. We select the number of folds to be equal to 3 to lower the execution time of the process. We define the number of iterations to be in total 3000. The range of different parameters equals to 4500000 different scenarios.
    

A note for transformation step: Normally, pipelines in scikit can accept specific functions on them (e.g. vectorizers or columntransformers). In order to include the additional features in our pipeline we used the scikit's function FunctionTransformer( ) which allows to pass user-defined functions to the pipeline.
